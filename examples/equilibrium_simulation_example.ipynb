# Ezample 6- Equilibroum simulation

### Import atmosphere model

```python
from pyaat.atmosphere import atmosISA
atm = atmosISA()
```

### Import gravity model

```python
from pyaat.gravity import VerticalConstant
grav = VerticalConstant()
```

### Import Aircraft model

```python
from pyaat.aircraft import Aircraft
airc = Aircraft()
```

### Import propulsion model

```python
from pyaat.propulsion import SimpleModel
prop = SimpleModel()
```

### Create a system

```python
from pyaat.system import system
Mysys = system(atmosphere = atm, propulsion = prop, aircraft = airc, gravity = grav)
```

# simulate cruize condition

### Compute equilibrium condition

```python
Xe, Ue = Mysys.trimmer(condition='cruize', HE = 5000., VE = 200)
```

### propagate equilibrium condition

```python
solution, control = Mysys.propagate(Xe, Ue, TF = 180, perturbation = False)
```

### Import plotter

```python
from pyaat.tools import plotter
pltr = plotter()
```

### Plot results

```python
pltr.states = solution
pltr.time = Mysys.time
pltr.control = control
```

```python
pltr.LinVel(frame = 'aero')
```

```python
pltr.LinVel(frame = 'body')
```

```python
pltr.LinPos()
```

```python
pltr.Attitude()
```

```python
pltr.AngVel()
```

```python
pltr.Controls()
```

```python
pltr.LinPos3D()
```

```python
pltr.linPos2D()
```
